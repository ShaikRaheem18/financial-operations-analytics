```python
import pandas as pd
import numpy as np

# Added 'r' before the path string
df_raw = pd.read_csv(r"C:\Users\rshie\financial-operations-analytics\data\raw\credit_card_transactions.csv")
df = df_raw.copy()

```

```python
columns_to_drop = [
    "Unnamed: 0",
    "first", "last",
    "street", "zip",
    "lat", "long", "merch_lat", "merch_long",
    "job", "gender",
    "unix_time", "merch_zipcode"
]

df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.shape
df.columns
```

```python
df["trans_date_trans_time"] = pd.to_datetime(df["trans_date_trans_time"])
```

```python
df.rename(columns={
    "trans_num": "transaction_id",
    "cc_num": "customer_id",
    "amt": "amount",
    "trans_date_trans_time": "transaction_date"
}, inplace=True)
```

```python
df = df[df["amount"] > 0]
```

```python
df["status"] = np.where(df["is_fraud"] == 1, "FAILED", "SUCCESS")
```

```python
df["dob"] = pd.to_datetime(df["dob"])

df["age"] = (pd.Timestamp("now") - df["dob"]).dt.days // 365

df["age_group"] = pd.cut(
    df["age"],
    bins=[18, 25, 35, 45, 55, 65, 100],
    labels=["18-25", "26-35", "36-45", "46-55", "56-65", "65+"]
)

df.drop(columns=["dob", "age"], inplace=True)
```

```python
df.isnull().sum()
df.describe()
```

```python
len(df_raw), len(df)
```

```python
final_columns = [
    "transaction_id",
    "customer_id",
    "transaction_date",
    "amount",
    "merchant",
    "category",
    "city",
    "state",
    "age_group",
    "status",
    "is_fraud"
]

df_final = df[final_columns]
```
